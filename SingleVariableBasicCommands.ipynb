```python
fgp_rounded = round(df["FG%"], 0) #Round values to X decimal places, use negative for nearest 1, 10, 100, etc...
fgp_rounded.head()
```

```python
#Import the book's helper libraries, these will help with visualizations 
import thinkstats2
import thinkplot
```

```python
df_hist = thinkstats2.Hist(fgp_rounded, label="Field Goal %")  #Create a Hist object - a building block of a histogram.
#The label part is optional.
```

```python
#The hist is basically a special data structure that the book author made.
#It basically just holds the same data as the dataframe, but it 'sets it up' to be graphed.
#It is more or less just the value_counts results like above. 
#We can take a peek at it
df_hist
```

```python
#Create a histogram for that Hist object. 
thinkplot.PrePlot(1) #Tell thinkplot how many things are about to get plotted
thinkplot.Hist(df_hist)
```

```python
#Add some options to the histogram
thinkplot.PrePlot(1) #Tell thinkplot how many things are about to get plotted
thinkplot.Hist(df_hist, color='purple') #Most 'normal' colors work
thinkplot.Config(xlim=[-5,105], ylim=[0,50])
```

```python
#Split some data and plot two histograms
#Split US people vs not US people - the Birth_Place column

#First, split the data into two dataframes. Print to check we did it
dfus = df[df["Birth_Place"] == "us"]
dfnon = df[df["Birth_Place"] != "us"]
#Filter out 0svalues, they aren't useful here
dfus = dfus[dfus["3P%"] > 1]
dfnon = dfnon[dfnon["3P%"] > 1]

#See what we did
print(dfus.head())
print(dfnon.head())
```

FYI - a few of the examples below use 3 point percentage. You don't really need to know what it means in basketball. The typical range is roughly from the low 30s to low 40s. 

```python
#Make the two hist objects
histUS = thinkstats2.Hist(dfus["3P%"], label="US Players") #Label is helpful with more than one histogram
histNon = thinkstats2.Hist(dfnon["3P%"], label="Furriners") 
```

```python
#Make the actual plots
thinkplot.PrePlot(2)
width = .5 #The width of the bars can be set, to make the graph nicer. Likely requires some guess and test
thinkplot.Hist(histUS, color="blue", width=width, align='left') #Color, width, and alignment add clarity
thinkplot.Hist(histNon, color="green", width=width, align='right')
thinkplot.Config()
```

```python
thinkplot.clf #Clear all the thinkplot stuff. Not usually needed.
```

```python
#Make a PMF - like a histogram, but: it is a line, and shows normalized percentages 
#instead of raw values
thinkplot.PrePlot(2)
thinkplot.Pmfs([histUS, histNon])
thinkplot.Config(style="line")
```

```python
#CDF - cumulative distribution, or running totals up to that point. 
df_no0 = round(df[df["3P%"] > 0], 0) #Take all values, remove 0s. Round to nearest %
df_cdf = thinkstats2.Cdf(df_no0["3P%"])
```

```python
#Create CDF plot
thinkplot.Cdf(df_cdf) #Make a cdf just like a hist
thinkplot.Show(xlabel="3 Point Percentage", ylabel="CDF")
```

```python
#What percentage of values is this value above?
#e.g. a 3P% of 37 is greater than what percentage of other values?
df_cdf.Prob(37)
```

```python
#Go the other way, to be greater than 73% of results, what value is that?
df_cdf.Value(.731)
```

```python
#Same-same with the percentile functions
df_cdf.PercentileRank(37)
```

```python
#Same-same with the percentile functions
df_cdf.Percentile(73.1)
```
